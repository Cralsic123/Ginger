```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('BETTER30.csv')
```

```python
df.sample(5)
```

```python
df = df.drop(columns = ["CONVERSATION_ID",	"CONVERSATION_STEP"])
```

```python
df = df.drop(columns = ["FEATURES",	"ANNOTATIONS", "CONTEXT"])
```

```python
df.head()
```

```python
df['LABEL'].value_counts()
```

```python
df['LABEL'] = df['LABEL'].replace({'neutral': 'not scam', 'legitimate': 'not scam', ' neutral': 'not scam', ' legitimate': 'not scam'})
df['LABEL'] = df['LABEL'].apply(lambda x: 'scam' if x not in ['not scam'] else x)
```

```python
df['LABEL'].value_counts()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
df['LABEL'] = encoder.fit_transform(df['LABEL'])
```

```python
import matplotlib.pyplot as plt
plt.pie(df['LABEL'].value_counts(), labels=['not spam','spam'],autopct="%0.2f")
plt.show()
```

```python
import nltk
```

```python
nltk.download('punkt')
```

```python
df['num_characters'] = df['TEXT'].apply(len)
```

```python
df.head()
```

**Tokenizing**

```python
df['num_words'] = df['TEXT'].apply(lambda x:len(nltk.word_tokenize(x)))
```

```python
df['num_characters'] = df['TEXT'].apply(len)
```

```python
df['num_sentences'] = df['TEXT'].apply(lambda x:len(nltk.sent_tokenize(x)))
```

```python
df.head()
```

```python
df[['num_characters','num_words','num_sentences']].describe()
```

```python
# for legitimate
df[df['LABEL'] == 0][['num_characters','num_words','num_sentences']].describe()
```

```python
# spam
df[df['LABEL'] == 1][['num_characters','num_words','num_sentences']].describe()
```

```python
import seaborn as sns
plt.figure(figsize=(12,6))
sns.histplot(df[df['LABEL'] == 0]['num_characters'])
sns.histplot(df[df['LABEL'] == 1]['num_characters'],color='red')
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df[df['LABEL'] == 0]['num_words'])
sns.histplot(df[df['LABEL'] == 1]['num_words'],color='red')
```
